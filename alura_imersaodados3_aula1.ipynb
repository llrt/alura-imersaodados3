## Análise univariada (i.e. de cada coluna isolada)

### Coluna _tratamento_

```python
dados['tratamento'].value_counts() # quais valores únicos existem na coluna?
```

```python
dados['tratamento'].value_counts(normalize=True) # qual a proporção entre os valores únicos encontrados
```

A coluna _tratamento_ é muito desbalanceada. Por que?

Será que a distribuição é mais homogênea quando olhamos em conjunto com outra coluna/variável?

```python
pd.crosstab(dados.tratamento, dados.tempo, margins=True) # gera uma tabela cruzando valores das colunas selecionadas
pd.crosstab(dados.tratamento, dados.tempo, normalize='columns') # gera uma tabela cruzando valores das colunas selecionadas, normalizando (i.e. obtendo a proporção)
```

Parece que _tratamento_ x _tempo_ mantém a mesma proporção geral de _tratamento_ ...

E _dose_ ?

```python
pd.crosstab(dados.tratamento, dados.dose, margins=True)
pd.crosstab(dados.tratamento, dados.dose, normalize='columns')
```

Também não.. _tratamento_ x _dose_ também mantém a mesma proporção..

E _droga_ ?

```python
pd.crosstab(dados.tratamento, dados.droga, margins=True) 
pd.crosstab(dados.tratamento, dados.droga, normalize='columns')
```

Bom, quando tem _droga_, não tem _tratamento_ = _'com_controle'_

```python
pd.crosstab(dados.droga, [dados.dose, dados.tratamento], margins=True) 
pd.crosstab(dados.droga, [dados.dose, dados.tratamento], normalize='columns')
```
